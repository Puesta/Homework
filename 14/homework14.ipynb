**Задание 1**

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
min_date_rate = ratings[['userId','timestamp']].groupby('userId').agg('min').reset_index()
max_date_rate = ratings[['userId','timestamp']].groupby('userId').agg('max').reset_index()
```

```python
min_date_rate.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
max_date_rate.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
full_date_rate = min_date_rate.merge(max_date_rate, how = 'left', on = 'userId')
full_date_rate['life'] = (full_date_rate['timestamp_max'] - full_date_rate['timestamp_min']).astype('timedelta64[m]')
```

```python
full_date_rate.head()
```

```python
# проверка на примере user 485
ratings.loc[ratings['userId'] == 485].sort_values(by='timestamp')
```

```python
full_date_rate.loc[full_date_rate['userId'] == 485]
```

```python
# проверка что не потерялись клиенты
len(ratings['userId'].unique()) == len(min_date_rate)
len(ratings['userId'].unique()) == len(max_date_rate)
```
